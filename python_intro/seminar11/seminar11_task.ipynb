```python
from sympy import *
from sympy.plotting import plot
```

```python
x = Symbol('x')
gx = -5*x**2+10*x-150
```

1. Находим корни

```python
[root1, root2] = solve(gx,x)
print(f'{root1}\n{root2}')
```

2. Находим интервалы, на которых функция возрастает

```python
solve(diff(gx)>0, x,check=False)
```

3. Находим интервалы, на которых функция убывает

```python
solve(diff(gx) < 0, x, check=False)
```

4. Строим график

```python
p = plot(gx)
```

5. Ищем вершину

```python
solve(diff(gx), x, check=False)
```

6. Определяем промежуток, на котором f > 0

```python
solve(gx > 0, x,check=False)
```

7. Определяем промежуток, на котором f < 0

```python
solve(gx < 0, x,check=False)
```
